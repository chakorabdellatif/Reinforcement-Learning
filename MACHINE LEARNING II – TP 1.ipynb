# Exercice 1 : Découverte et Exploration d’un Environnement Gym

```python
print(f"Espace d’actions : {env.action_space}")  
print(f"Espace d’observations : {env.observation_space}")  

for _ in range(100):  
    action = env.action_space.sample()  
    observation, reward, done, _, _ = env.step(action)  
    print(f"Action : {action}, Observation : {observation}, Reward : {reward}")  
    if done:  
        env.reset()  
env.close()
```

#  Exercice 2 : Manipulation des Observations et Récompenses

```python
# prendre une seule action
action = env.action_space.sample()  # Action aléatoire (0 ou 1)
observation, reward, done, _, _ = env.step(action)

# affichage des valeurs retournées
print(f"Action prise : {action}") 
print(f"Observation : {observation}") 
print(f"Récompense : {reward}") 
print(f"Épisode terminé ? {done}") 

env.close()
```

```python
# effectuer plusieurs essais pour observer les variations
for i in range(5): 
    action = env.action_space.sample()
    observation, reward, done, _, _ = env.step(action)
    
    print(f"\nEssai {i+1}:")
    print(f"Action prise : {action}")
    print(f"Observation : {observation}")
    print(f"Récompense : {reward}")
    print(f"Épisode terminé ? {done}")
    
    if done:
        env.reset()

env.close()
```

# Exercice 3 : Contrôle Manuel de l’Agent

```python
# demander à l’utilisateur d’entrer une action (0 ou 1)
while True:
    try:
        action = int(input("Entrez une action (0 = gauche, 1 = droite) : "))
        if action in [0, 1]:
            break
        else:
            print("Veuillez entrer 0 ou 1.")
    except ValueError:
        print("Veuillez entrer un nombre valide.")
```

```python
# exécuter l’action dans l’environnement et afficher les nouvelles observations
observation, reward, done, _, _ = env.step(action)

print(f"Action prise : {action}")
print(f"Nouvelle Observation : {observation}")
print(f"Récompense : {reward}")
print(f"Épisode terminé ? {done}")
```

```python
# répéter l’opération jusqu’à la fin de l’épisode
i = 0 
done = False
while not done:
    while True:
        try:
            action = int(input("Entrez une action (0 = gauche, 1 = droite) : "))
            if action in [0, 1]:
                break
            else:
                print("Veuillez entrer 0 ou 1.")
        except ValueError:
            print("Veuillez entrer un nombre valide.")
    
    observation, reward, done, _, _ = env.step(action)
    i +=1

    
    print(f"Nouvelle Observation : {observation}")
    print(f"Récompense : {reward}")
    print(f"Épisode terminé ? {done}")
print(f"la durée totale de l’épisode : {i} actions")
env.close()
```

# Exercice 4 : Évaluation des Performances d’une Politique Aléatoire

```python
duree = []
for episode in range(5):
    observation, _ = env.reset()
    done = False
    steps = 0

    while not done:
        action = env.action_space.sample()
        observation, reward, done, _, _ = env.step(action)
        steps += 1

    duree.append(steps)  
    print(f"Épisode {episode+1} terminé en {steps} actions.")

# calculer la durée moyenne
moyenne_duree = sum(duree) / 5
print(f"durée moyenne d’un épisode : {moyenne_duree:.2f} actions.")

env.close()
```
